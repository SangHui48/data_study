## 4.3 수치형 데이터 시각화

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### 4.3.1 히스토그램(hisplot)

* **히스토그램**(histogram)은 수치형 데이터의 구간별 빈도수를 나타내는 그래프이다. histplot()함수로 그릴수 있다.

```python
sns.histplot(data=titanic, x='age')
```

```python
# 구간 수 설정
sns.histplot(data=titanic, x='age', bins=10)
```

```python
# 범주를 나눠서 보기
sns.histplot(data=titanic, x='age', hue='alive')
```

```python
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack')
```

### 4.3.2 커널밀도추정 함수 그래프(kdeplot)

* **커널밀도추정**(kernal density estimation) 함수는 쉽게 생각해 히스토그램을 매끄럽게 곡선으로 연결한 그래프 정도로 이해하면 된다.

```python
sns.kdeplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack')
```

### 4.3.3 분포도(displot)

* **분포도**는 수치형 데이터 하나의 분포를 나타내는 그래프이다.

```python
sns.displot(data=titanic, x='age')
```

```python
# 커널밀도추정 함수로 바꿔보기
sns.displot(data=titanic, x='age', kind='kde')
```

```python
sns.displot(data=titanic, x='age', kde=True) #동시에 그리기
```

### 4.3.4 러그플롯(rugplot)

* **러그 플롯**은 **주변 분포**(marginal distribution)를 나타내는 그래프이다. 단독으로 사용하기보다는 주로 다른 분포도 그래프와 함께 사용한다.

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age')
```

## 4.4 범주형 데이터 시각화

### 4.4.1 막대 그래프(barplot)

* **막대 그래프**는 범주형 데이터 값에 따라 수치형 데이터 값이 어떻게 달라지는지 파악할 떄 사용한다.
* 보통 일반적으로 x파라미터에 범주형 데이터, y파라미터에 수치형 데이터를 전달한다.

```python
sns.barplot(x='class', y='fare', data=titanic)
```

### 4.4.2 포인트플롯(pointplot)

* **포인트 플롯**은 막대 그래프와 모양만 다를 뿐 동일한 정보를 제공한다.

```python
sns.pointplot(x='class', y='fare', data=titanic)
```

* 포인트 플롯을 사용하면 좋을때 : 한 화면에 여러 그래프를 그려야 할 떄

### 4.4.3 박스플롯(boxplot)

* **박스 플롯**(box plot)은 막대 그래프나 포인트플롯보다 더 많은 정보를, 구체적으로 5가지 요약 수치를 제공한다. 5가지 요약수치는 최솟값과 사분위 수
* 제 1사분위 수(Q1) : 전체 데이터 중 하위 25%에 해당하는 값
* 제 2사분위 수(Q2) : 50%에 해당하는 값(중앙값)
* 제 3사분위 수(Q3) : 상위 25%에 해당하는 값
* 사분위 범위 수(IQR) : Q3 - Q1
* 최댓값 : Q3 + (1.5 * IQR)
* 최솟값 : Q1 - (1.5 * IQR)

```python
sns.boxplot(x='class', y='age', data=titanic)
```

### 4.4.4 바이올린플롯(violinplot)

* **바이올린플롯**은 박스플롯과 커널밀도추정 함수 그래프를 합쳐 놓은 그래프라고 볼 수 있다.

```python
sns.violinplot(x='class', y='age', data=titanic)
```

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True)
```

### 4.4.5 카운트플롯(countplot)

* **카운트 플롯**은 범주형 데이터의 개수를 확인할 때 사용되는 그래프이다. 주로 범주형 피처나 범주형 타깃값의 분포가 어떤지 파악하는 용도로 사용한다.

```python
sns.countplot(x='class', data=titanic)
```

```python
# y로 바꾸면 가로로 누운 형태의 그래프가 된다.
sns.countplot(y='class', data=titanic)
```

### 4.4.6 파이 그래프(pie)

* **파이 그래프**는 범주형 데이텅별 비율을 알아볼 때 사용하기 좋다. seaborn은 불가능 matplotlib은 가능

```python
import matplotlib.pyplot as plt

x = [10, 60, 30]
labels = ['A', 'B', 'C']
plt.pie(x=x, labels=labels, autopct='%.1f%%')
```

## 4.5 데이터 관계 시각화

### 4.5.1 히트맵(heatmap)

* **히트맵**은 데이터 간 관계를 색상으로 표현한 그래프이다. 비교해야 할 데이터가 많을 때 주로 사용한다.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
```

```python
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

### 4.5.2 라인플롯(lineplot)

* **라인 플롯**은 두 수치형 데이터 사이의 관계를 나타낼 때 사용한다.

```python
sns.lineplot(x='year', y='passengers', data=flights)
```

### 4.5.3 산점도(scatterplot)

* **산점도**는 두 데이터의 관계를 점으로 표현하는 그래프이다.

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
```

### 4.5.4 회귀선을 포함한 산점도 그래프(regplot)

* 산점도와 회귀선을 동시에 그려주는 함수. 회귀선을 그리면 전반적인 상관관계 파악이 좀 더 쉽다.

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
sns.regplot(x='total_bill', y='tip', ci=99, data=tips) #신뢰구간 99% 늘리기
```
