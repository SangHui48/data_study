* 피처 요약표

```python
def resumetable(df):
    print(f'데이터셋 형상: {df.shape}')
    summary = pd.DataFrame(df.dtypes, columns=['데이터 타입'])
    summary['결측값 개수'] = (df == -1).sum().values
    summary['고윳값 개수'] = df.nunique().values
    summary['데이터 종류'] = None
    for col in df.columns:
        if 'bin' in col or col == 'target':
            summary.loc[col, '데이터 종류'] = '이진형'
        elif 'cat' in col:
            summary.loc[col, '데이터 종류'] = '명목형'
        elif df[col].dtype == float:
            summary.loc[col, '데이터 종류'] = '연속형'
        elif df[col].dtype == int:
            summary.loc[col, '데이터 종류'] = '순서형'
            
    return summary
```

```python
summary = resumetable(train)
summary
```

### 8.2.2 데이터 시각화

* 타깃값 분포를 활용해 타깃값이 얼마나 불균형한지 먼저 알아보기

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

* 타깃값 분포

```python
def write_percent(ax, total_size):
    for patch in ax.patches:
        height = patch.get_height()
        width = patch.get_width()
        left_coord = patch.get_x()
        percent = height/total_size*100
        
        ax.text(left_coord + width/2.0, height + total_size*0.001, '{:1.1f}%'.format(percent), ha='center')

mpl.rc('font', size=15)
plt.figure(figsize=(7,6))

ax = sns.countplot(x='target', data=train)
write_percent(ax, len(train))
ax.set_title('Target Distribution')
```

* **통계적 유효성**은 barplot()을 그릴 때 나타나는 신뢰구간으로 판단한다. 신뢰구간이 좁다면 통계적으로 어느 정도 유효하다고 보고, 구간이 넓다면 신뢰하기 어렵다 판단한다. 예를 들어서 '임꺽정의 키가 170-175cm 정도 돼' 라는 말이 '임꺽정의 키는 150-200cm 정도 돼'라는 말보다 더 구체적이고 신뢰가는 것과 같은 이치이다.

* 이진 피처

```python
import matplotlib.gridspec as gridspec

def plot_target_ratio_by_features(df, features, num_rows, num_cols, size=(12, 18)):
    mpl.rc('font', size=9)
    plt.figure(figsize=size)
    grid = gridspec.GridSpec(num_rows, num_cols)
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    
    for idx, feature in enumerate(features):
        ax = plt.subplot(grid[idx])
        sns.barplot(x=feature, y='target', data=df, palette='Set2', ax=ax)
```

```python
bin_features = summary[summary['데이터 종류'] == '이진형'].index
plot_target_ratio_by_features(train, bin_features, 6, 3)
```

* 명목형 피처

```python
nom_features = summary[summary['데이터 종류'] == '명목형'].index

plot_target_ratio_by_features(train, nom_features, 7, 2)
```

* 순서형 피처

```python
ord_features = summary[summary['데이터 종류'] == '순서형'].index

plot_target_ratio_by_features(train, ord_features, 8, 2, (12, 20))
```

* 연속형 피처

```python
cont_features = summary[summary['데이터 종류'] == '연속형'].index

plt.figure(figsize=(12,16))
grid = gridspec.GridSpec(5, 2)
plt.subplots_adjust(wspace=0.2, hspace=0.4)

for idx, cont_feature in enumerate(cont_features):
    # 값을 5개의 구간으로 나누기
    train[cont_feature] = pd.cut(train[cont_feature], 5)
    
    ax = plt.subplot(grid[idx])
    sns.barplot(x=cont_feature, y='target', data=train, palette='Set2', ax=ax)
    ax.tick_params(axis='x', labelrotation=10)
```

* 연속형 피처2

* 피어슨 상관관계를 이용해 피처를 제거해 본다. 만약 두 피처의 상관관계가 높다면 둘 중 하나는 제거를 해주는게 좋다. 상관관계가 강하면 타깃값의 예측력도 비슷하기 때문이다. 그런 피처가 있다면 모델의 성능이 떨어질수도 있다.  
    0.00 - 0.19 : 아주 약함
    0.20 - 0.39 : 약함
    0.40 - 0.59 : 보통
    0.60 - 0.79 : 강함
    0.80 - 1.0 : 아주 강함

### 분석 정리 및 모델링 전략

* **분석 정리**
    * 데이터가 크고 피처 수도 많다.
    * 피처명만으로 분류별 혹은 데이터 종류별 피처들을 구분해 추출해낼 수 있다.
    * **결측값 처리** : 결측값 자체에 타깃값 예측력이 있다면 고윳값으로 간주한다.
    * **결측값 처리** : 피처 간 상관관계 분석은 결측값 제거 후 수행한다.
    * **피처 제거** : 신뢰구간이 넓으면 통계적 유효성이 떨어져 믿을 수 없다.
    * **피처 제거** : 고윳값별 타깃값 비율에 차이가 없다면 타깃값 예측력이 없다.
    * **피처 제거** : (연속형 데이터의 경우) 구간별 타깃값 차이가 거의 없다면 타깃값 예측력이 없다.
    * **피처 제거** : 일반적으로 강한 상관관계를 보이는 두 피처가 있으면 둘 중 하나를 제거하는게 좋다.
    
* **모델링 전략**
    * **베이스라인 모델** : LightGBM
        * 훈련 및 예측 : OOF 예측(과대적합 방지 + 앙상블 효과)
    * 성능 개선1 : LightGBM 유지
        * **피처 엔지니어링**: 파생 피처 추가
        * **하이퍼파라미터 최적화** : 베이지안 최적화
    * 성능 개선2 : XGBoost (모델만 변경)
    * 성능 개선3 : LightGBM + XGBoost 앙상블
