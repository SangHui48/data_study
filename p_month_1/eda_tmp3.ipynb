```python
import pandas as pd
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_2015 = pd.read_csv('data/2015.csv')
df_2016 = pd.read_csv('data/2016.csv')
```

```python
df_list = [df_2015, df_2016]
yr_list = ['2015', '2016']
```

```python
font_path = "C:/Users/Shin/AppData/Local/Microsoft/Windows/Fonts/KOPUBWORLD DOTUM MEDIUM.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
```

```python
for i in range(len(df_list)):
    fig, ax = plt.subplots(figsize=(16,10))
    sns.heatmap(df_list[i].corr(), # 위에서 전처리한 데이터프레임
                annot = True, # 숫자 표시 여부
                ax = ax, # 히트맵을 그릴 격자
                linewidths = 0.4, # 선의 굵기
                linecolor = 'white', # 선의 색깔
                fmt = '.2f', # 소수점 포맷팅 형태
                cmap = 'YlGnBu') # colormap 형태

    plt.title('Correlation - Year: ' + yr_list[i], size = 20, pad=20)
    plt.ylabel('', size = 12)

    plt.xticks(rotation=60, size=14)
    plt.yticks(size=14)
    plt.tight_layout()
    plt.show()
```

연도별 상관계수 확인 결과
1) 2015년: Economy(GDP for Capita)(-0.79) → Health(Life Expectancy)(-0.74) → Family(-0.73) 순
2) 2016년: Economy(GDP for Capita)(-0.79) → Health(Life Expectancy)(-0.77) → Family(-0.73) 순  

연도별 순위는 동일하며, 상관계수 크기도 거의 비슷함  
⇒ 2015년과 2016년의 행복지수에 대해서 큰 영향을 끼치는 요인은 세 가지로 추려낼 수 있음

산점도를 확인해본 결과, 위의 세 가지 요인과 Happiness Rank는 높은 음의 상관관계를 보임.  
반대로 Happiness Score는 높은 양의 상관관계를 보인다.

```python
plt.figure(figsize=(16,4))
sns.barplot(x=df_2015.nsmallest(30, 'Happiness Rank')['Country'], y=df_2015.nsmallest(30, 'Happiness Rank')['Happiness Score'])
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.barplot(x=df_2016.nsmallest(30, 'Happiness Rank')['Country'], y=df_2016.nsmallest(30, 'Happiness Rank')['Happiness Score'])
plt.xticks(rotation=60)
plt.show()
```

2015년과 2016년 모두 주로 북유럽 국가들이 높은 행복 지수를 보임.  
Canada 역시 행복 지수가 높은 편.

```python
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_all = pd.concat([df_2015, df_2016])
```

```python
df_all.head(5)
```

```python
df_all.tail(5)
```

```python
df_all[['Year','Happiness Score']].groupby(['Year']).mean()
```

```python
hs_by_yr = df_all['Happiness Score'].groupby(df_all['Year']).mean()
plt.figure(figsize=(4,4))
ax = sns.barplot(x=hs_by_yr.index, y=hs_by_yr.values)
ax.bar_label(ax.containers[0])
```

연도별 평균 행복 지수는 2016년이 2015년보다 근소하게 높다.

```python
df_all[['Region','Happiness Score']].groupby(['Region']).mean()
```

```python
hs_by_rg = df_all['Happiness Score'].groupby(df_all['Region']).mean()
plt.figure(figsize=(16,4))
ax = sns.barplot(x=hs_by_rg.index, y=hs_by_rg.values)
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45)
plt.show()
```

지역별 평균 행복 지수는 Australia and New Zealand, North America, Western Europe 순으로 높음.  
낮은 순으로는 Sub-Saharan Africa, Southern Asia, Southeastern Asia.
