```python
import os
```

```python
path = '/content/drive/MyDrive/data/programmers_ai/yujatea/comp_3/dataset/'
```

```python
import pandas as pd

train_df = pd.read_csv(os.path.join(path, 'my_train.csv'))
test_df = pd.read_csv(os.path.join(path, 'my_test.csv'))
```

```python
train_df.head()
```

```python
train_df['box_area'] = train_df['w'] * train_df['h']
train_df.head()
```

```python
test_df['box_area'] = test_df['w'] * test_df['h']
```

```python
test_df.head()
```

```python
train_df.to_csv(os.path.join(path,'my_train_df.csv'), index=False)
test_df.to_csv(os.path.join(path,'my_test_df.csv'), index=False)
```

```python
info_dict = {"info": {
    "year":"2022","version":"1",
    "description":"Exported from roboflow.ai",
    "contributor":"",
    "url":"https://public.roboflow.ai/object-detection/undefined","date_created":"2022-12-17T02:13:57+00:00"
}}

license_dict = {
    "licenses":[{"id":1,"url":"https://creativecommons.org/licenses/by/4.0/","name":"CC BY 4.0"}]
}

categories_dict = {
        "categories":[
        {"id": int(0), "name":"Water buffalo","supercategory":"animal"},
        {"id":int(1), "name":"Elephant","supercategory":"animal"},
        {"id":int(2), "name":"Rhinoceros","supercategory":"animal"},
        {"id":int(3), "name":"Zebra","supercategory":"animal"}
    ]
}
```

```python
images_dict = {
    "images": []
}

annotations_dict = {
    "annotations": []
}
```

```python
from tqdm.notebook import tqdm

image_set = set()

for idx in tqdm(range(len(train_df))):
    tmp_series = train_df.iloc[idx]
    if int(tmp_series['filename'][:-4]) in image_set:
        continue
    else:
        tmp_image_dict = {
        "id": int(tmp_series['filename'][:-4]),
        "file_name": tmp_series['filename'],
        "height": int(tmp_series['height']),
        "width": int(tmp_series['width'])
        }
        images_dict["images"].append(tmp_image_dict)
    
    tmp_annotation_dict = {
        "id": idx,
        "image_id": int(tmp_series['filename'][:-4]),
        "category_id": int(tmp_series['label']),
        "bbox":[
            int(tmp_series['x']),
            int(tmp_series['y']),
            int(tmp_series['w']),
            int(tmp_series['h'])
        ],
        "area": int(tmp_series['box_area']),
    }
    annotations_dict["annotations"].append(tmp_annotation_dict)
```

```python
from tqdm.notebook import tqdm

image_set = set()

for idx in tqdm(range(len(test_df))):
    tmp_series = test_df.iloc[idx]
    if int(tmp_series['filename'][:-4]) in image_set:
        continue
    else:
        tmp_image_dict = {
        "id": int(tmp_series['filename'][:-4]),
        "file_name": tmp_series['filename'],
        "height": int(tmp_series['height']),
        "width": int(tmp_series['width'])
        }
        images_dict["images"].append(tmp_image_dict)
    
    tmp_annotation_dict = {
        "id": idx,
        "image_id": int(tmp_series['filename'][:-4]),
        "category_id": int(tmp_series['label']),
        "bbox":[
            int(tmp_series['x']),
            int(tmp_series['y']),
            int(tmp_series['w']),
            int(tmp_series['h'])
        ],
        "area": int(tmp_series['box_area']),
    }
    annotations_dict["annotations"].append(tmp_annotation_dict)
```

```python
annotation_all_dict = {}
annotation_all_dict.update(info_dict)
annotation_all_dict.update(license_dict)
annotation_all_dict.update(categories_dict)
annotation_all_dict.update(images_dict)
annotation_all_dict.update(annotations_dict)
```

```python
import json

with open(os.path.join(path, "_annotations.coco.json"), "w") as json_file:
    json.dump(annotation_all_dict, json_file)
```
