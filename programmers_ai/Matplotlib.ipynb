```python
import numpy as np
import pandas as pda
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4,5]) # 꺾은선 그래프. 실제 plotting을 하는 함수 y=x+1
plt.show() #plt를 확인하는 명령
```

```python
plt.figure(figsize=(6,6))
plt.plot([0,1,2,3,4,5]) # 꺾은선 그래프. 실제 plotting을 하는 함수 y=x
plt.show() #plt를 확인하는 명령
```

```python
# numpy.array를 이용해서 함수 그래프 그리기
# y = x^2
x = np.arange(6)
y = np.square(x)

plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-10, 10, 0.01) # c:0.01
plt.plot(x, x**2)
plt.show()
```

```python
# x,y축에 설명 추가하기
x = np.arange(-10, 10, 0.01)
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.plot(x, x**2)
plt.show()
```

```python
# x,y축 범위 설정하기
x = np.arange(-10, 10, 0.01)
plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]

plt.plot(x, x**2)
plt.show()
```

```python
# x,y축 눈금 설정하기
x = np.arange(-10, 10, 0.01)
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 26, 5)])

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기
x = np.arange(-10, 10, 0.01)
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 26, 5)])

plt.title("y=x^2 graph")

plt.plot(x, x**2)
plt.show()
```

```python
x = np.arange(-10, 10, 0.01)
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 26, 5)])
plt.title("y=x^2 graph")

plt.plot(x, x**2, label='trend') #label 추가
plt.legend()
plt.show()
```

```python
x = np.arange(20) # 0~ 19
y = np.random.randint(0, 20, 20) # 0~20까지 난수 20번 생성

plt.xlabel('x value')
plt.ylabel('y value')
plt.axis([0, 21, 0, 21])
plt.yticks([i for i in range(0, 21, 5)])
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(20) # 0~ 19
y = np.random.randint(0, 20, 20) # 0~20까지 난수 20번 생성

plt.xlabel('x value')
plt.ylabel('y value')
plt.axis([0, 21, 0, 21])
plt.yticks([i for i in range(0, 21, 5)])
plt.scatter(x, y)
plt.show()
```

```python
plt.title('Box plot of y')
plt.boxplot(y)
plt.show()
```

```python
plt.bar(x, y)
plt.xticks([i for i in range(0, 21)])
plt.show()
```

```python
vote_df = pd.DataFrame({"name": ['Andy', 'Bob', 'Cat'], "vote": [True, True, False]})

vote_df
```

```python
# in matplotlib barplot
vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
# sns의 countplot
sns.countplot(x=vote_df['vote'])
plt.show()
```
